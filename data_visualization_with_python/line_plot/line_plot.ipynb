# Line Plot

A line graph is commonly used to display change over time as a series of data points connected by straight line segments on two axes. The line graph therefore helps to determine the relationship between two sets of values, with one data set always being dependent on the other set.

**import necessary pacakages**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
sns.set_style('darkgrid')
print("Setup Complete")
```

**lets load the covid-19 dataset**

```python
df=pd.read_csv('covid-19_data.csv')
```

```python
df.head()
```

```python
df.shape
```

Lets draw line plot

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='ObservationDate',y='Confirmed',data=df,marker='o')
sns.lineplot(x='ObservationDate',y='Deaths',data=df,marker='D')
sns.lineplot(x='ObservationDate',y='Recovered',data=df,marker='^')
plt.legend(['Confirmed','Deaths','Recovered'])
plt.xticks(rotation=90)
plt.show()
```

**An easy way !!!!**

```python
df.set_index('ObservationDate',inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df,markers=['o','D','^'])


plt.xticks(rotation=90)
plt.show()
```
