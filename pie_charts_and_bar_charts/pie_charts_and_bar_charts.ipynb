# Pie charts and Bar charts

**What are pie charts and bar charts**
Pie charts represent data in a circle, with “slices” corresponding to percentages of the whole, whereas bar graphs use bars of different lengths to represent data in a more flexible way

**Why do we use pie charts and bar charts**
<ul>
    <li>To analyse categorical data</li>
    <li>To compare different types of values</li>
</ul>

Though pie charts and bar charts both used for the same purposes but now a days bar chart is more preferrble over pie chart.
I will explain along the way in this tutorial why bar charts are more preferable. So les's start

**As always lets import the necessary pacakages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
new_df=pd.read_csv('country_spread.csv')
```

```python
new_df
```

# Pie chart

```python
plt.figure(figsize=(10,10))
plt.pie(new_df.total,labels=new_df.country,
        autopct='%1.1f%%',
        explode=[0.5,0.3,0,0,0.1,0],
        pctdistance=0.8,
        colors=['#FE666F','#FE66D7','#AB66FE','#6689FE','#66FEB0','#F7FE66'],
        startangle=90)
plt.legend(labels=new_df.country)
plt.show()
```

# Bar chart

```python
plt.figure(figsize=(10,8))
sns.barplot(x=new_df.country,y=new_df.total)
plt.show()
```
