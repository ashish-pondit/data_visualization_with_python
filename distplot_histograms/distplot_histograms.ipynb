# Distplot (Histograms)

**What is histograms**

A **histogram** is an approximate representation of the distribution of numerical or categorical data.

**Why use histograms?**
<ol>
    <li>
        Histograms are used to show <b>distributions</b> of variables
    </li>
    <li>
        To plot a <b>univariate</b> distribution of observations.
    </li>
    <li>
        To show the underlying <b>frequency distribution</b> of the variable.
    </li>
</ol>


**Let's import all the necessary pacakages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
print('setup complete')
```

```python
house_df=pd.read_csv('house_price.csv')
```

```python
house_df.head()
```

```python
from scipy.stats import norm
```

```python
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'])


plt.show()
```

```python
#edit bin parameter
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],bins=10)


plt.show()
```

```python
#draw only kde plot
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],hist=False)


plt.show()
```

```python
#draw only histograms
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],kde=False)


plt.show()
```

```python
#add frequency line
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],rug=True)


plt.show()
```

```python
#use gaussian distribution instead of kde
from scipy.stats import norm
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],rug=True,fit=norm,kde=False)


plt.show()
```

```python
#draw the figure on y axis
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],rug=True,vertical=True)


plt.show()
```

```python
#change color
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],rug=True,color='y')


plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.distplot(house_df['SalePrice'],hist=True,rug=True,
            rug_kws={'color':'b'},
            kde_kws={'color':'g','lw':3},
            hist_kws={'color':'r','histtype':'step','linewidth':3,'alpha':1})
plt.show()
```
